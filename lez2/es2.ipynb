```python
import numpy as np
import random 
import matplotlib.pyplot as plt
import math 
```

 $p(y)=(dy/dx)^-1p(x)$

```python
def f(x):
    p_x=1/(10-0.1)
    return p_x*math.log(10)*10**x
x_teorico = np.linspace(-1, 1, 100)   # Genera 100 valori di x tra -5 e 5
y_teorico = f(x_teorico)
plt.plot(x_teorico, y_teorico)
plt.title('Grafico della funzione')
plt.show()
```

```python
# Genero 10.000 punti distribuiti uniformemente tra 0.1 e 10
x = np.random.uniform(low=0.1, high=10, size=100000)
# Plotto i dati in un istogramma
plt.hist(x, bins=50)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Distribuzione uniforme tra 0.1 e 10')
plt.show()
```

```python
y=[]
for i in range (0, np.size(x)):
    y.append(math.log10(x[i]))
```

```python
plt.hist(y, bins=100,density=True)
plt.xlabel('y')
plt.ylabel('p(y)')
plt.title('Distribuzione logaritmica tra -1 e 1')

plt.show()

```

```python
fig, ax = plt.subplots()
ax.plot(x_teorico, y_teorico, color='red', label='pdf teorica')
ax.hist(y, bins=100, alpha=0.5, label='Istogramma di y',density=True)
ax.legend(loc='best')
ax.set_xlabel('y')
ax.set_ylabel('p(y)')
ax.set_title('istogramma e pdf teorica di y')

# Mostrare il grafico
plt.show()
```

```python
log_mean_x=math.log10(np.mean(x))
mean_y=np.mean(y)
print('logaritmo della media di x:',log_mean_x,'\n')
print('media di y:',mean_y)
```

```python
log_median_x=math.log10(np.median(x))
median_y=np.median(y)
print('logaritmo della media di x:',log_median_x,'\n')
print('media di y:',median_y)
```

```python
fig, ax = plt.subplots()
ax.plot(x_teorico, y_teorico, color='red', label='pdf teorica')
ax.hist(y, bins=100, alpha=0.5, label='Istogramma di y',density=True)
ax.legend(loc='best')
ax.set_xlabel('y')
ax.set_ylabel('p(y)')
ax.set_title('istogramma e pdf teorica di y')
ax.axvline(x=log_mean_x, color='blue', linestyle='--',label='media delle due distribuzioni')
ax.axvline(x=mean_y, color='blue', linestyle='--')
# Mostrare il grafico
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_teorico, y_teorico, color='red', label='pdf teorica')
ax.hist(y, bins=100, alpha=0.5, label='Istogramma di y',density=True)
ax.legend(loc='best')
ax.set_xlabel('y')
ax.set_ylabel('p(y)')
ax.set_title('istogramma e pdf teorica di y')
ax.axvline(x=log_median_x, color='blue', linestyle='--',label='mediane delle due distribuzioni')
ax.axvline(x=median_y, color='blue', linestyle='--')
# Mostrare il grafico
ax.legend()
plt.show()
```
